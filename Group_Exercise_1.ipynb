### 1.3. Create Lead() and Lag()

```python
#def lead(ar,x):
#    return np.append(ar,np.array([np.nan for i in range(x)]))[x:]

def lead(x, n):

    y = []
    for i in range(0, len(x)):

        x_new = x[i,n:]
        NaN_Array = np.full(n, np.nan)
        a = np.concatenate((x_new, NaN_Array), axis= 0)
        y.append([a])
        
    y = np.asarray(y)

    return y
```

```python
# example
lead(x_a,2)
```

```python
#def lag(ar,x):
#    return np.append(np.array([np.nan for i in range(x)]),ar)[:ar.size-x]

def lag (x, n):

    y = []
    for i in range(0, len(x)):
        
        x_new = x[i,:-n]
        NaN_Array = np.full(n, np.nan)
        a = np.concatenate((NaN_Array, x_new), axis= 0)
        y.append([a])

    y = np.asarray(y)
    return y
```

```python
# example
lag(x_a,2)
```

### 1.4. Pairwise Distance

```python
# generate random data
n = 10
d = 5

x_li = [[np.random.normal() for i in range(n)] for j in range(d)]
x_ar = np.array(x_li)

#y_li = [[np.random.normal() for i in range(n)]]*d
y_li = [np.random.normal() for i in range(n)]
y_ar = np.array(y_li)
```

```python
# size of the arrays
#x_ar.shape, y_ar.shape
x_ar,y_ar
```

```python
pairwisedist = []
y2 =  np.dot(y_ar, y_ar, out=None)
for point in range(d):
    sup = np.dot(x_ar[point], x_ar[point], out=None) + y2 - 2*np.dot(x_ar[point], y_ar, out=None)
    pairwisedist.append(sup)
answer4 = pairwisedist
answer4
```

### 2.2.1. Select all columns between Year and Day (Inclusive)

```python
# select file
path2 = "C:/Users/drr19/OneDrive/Bocconi - DSBA/1 - Statistics/ProbabilityAndStatistics/Assignment_1/data/nycflights13/nycflights13_flights.csv"
flights = pd.DataFrame(pd.read_csv(path2, sep=",", header=54, encoding="utf-8"))
flights.head()
```

```python
# Define a funcion: Output as Binary List
def RangeYearDate(x,between=1):
    
    # between = 1 for columns between Year and Date
    # between = 0 for all other columns
    
    selected = []
    activate = (1-between)
    
    for keyname in list(x.keys()):
        if keyname == "year": activate = between
        if activate == between: selected.append(between)
        if activate == (1-between): selected.append((1-between))
        if keyname == "day": activate = (1-between)
            
    return selected

# Define a funcion: Output as DataFrame of selected Columns

def selectcolumns(df,lista):
    output = pd.DataFrame()
    for i in range(len(lista)):
        if lista[i] == 1:
            output[df.keys()[i]] = df[df.keys()[i]]
    return output
```

```python
# Output as List of Binary Columns
selection_between = RangeYearDate(flights,between=1)
print(selection_between)

# Output as DataFrame
selectcolumns(flights,selection_between)
```

### 2.2.2. Select all other columns

```python
# Output as List of Binary Columns
selection_others = RangeYearDate(flights,between=0)
print(selection_others)

# Output as DataFrame
selectcolumns(flights,selection_others)
```
